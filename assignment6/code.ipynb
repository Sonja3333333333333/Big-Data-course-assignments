```python
import pandas as pd
from scipy import stats
```

Використовуємо дані з файлів bda_dataset6_2.xlsx

```python
data = pd.read_excel("bda_dataset6_2.xlsx")
data
```

Видаляємо рядки із пропущеними значеннями

```python
data.dropna(inplace=True)
data
```

```python
control_campaign_data = data[data["Campaign Name"] =="Control Campaign"]
control_campaign_data
```

```python
test_campaign_data = data[data["Campaign Name"] =="Test Campaign"].reset_index(drop=True)
test_campaign_data
```

### Тестування середньої кількості кліків

```python
clicks_control_campaign = control_campaign_data["# of Website Clicks"]
clicks_test_campaign = test_campaign_data["# of Website Clicks"]

t_statistic, p_value = stats.ttest_ind(clicks_control_campaign, clicks_test_campaign)


print("T-статистика:", t_statistic)
print("p-value:", p_value)
```

Від’ємне t-значення вказує, що середнє число кліків у контрольній кампанії менше, ніж у тестовій. Проте p-значення = 0.12 > 0.05, що свідчить про недостатні підстави для відхилення нульової гіпотези про відсутність різниці. Ми не виявили статистично значущої різниці у кліках на вебсайті між двома групами.

### Тестування середньої кількості пошукових запитів

```python
searches_control_campaign = control_campaign_data["# of Searches"]
searches_test_campaign = test_campaign_data["# of Searches"]

t_statistic, p_value = stats.ttest_ind(searches_control_campaign, searches_test_campaign)

print("T-статистика:", t_statistic)
print("p-value:", p_value)
```

Від’ємне t-значення вказує, що середня кількість пошукових запитів у контрольній кампанії менша, ніж у тестовій. Проте p-значення = 0.26 > 0.05, що свідчить про недостатні підстави для відхилення нульової гіпотези про відсутність різниці. Ми не виявили статистично значущої різниці у кількості пошукових запитів між двома групами.

### Тестування середньої кількості додавань до кошика

```python
add_to_cart_control_campaign = control_campaign_data["# of Add to Cart"]
add_to_car_test_campaign = test_campaign_data["# of Add to Cart"]

t_statistic, p_value = stats.ttest_ind(add_to_cart_control_campaign , add_to_car_test_campaign)

print("T-статистика:", t_statistic)
print("p-value:", f"{p_value:.10f}")
```

Високе додатне t-значення вказує, що середня кількість додавань до кошика у контрольній кампанії суттєво більша, ніж у тестовій.
p-значення = 0.00008 < 0.05 є дуже малим, що свідчить про сильні підстави для відхилення нульової гіпотези про відсутність різниці.
Різниця між групами є статистично значущою і з дуже малою ймовірністю могла виникнути випадково.

### Чи ефективною була експериментальна маркетингова кампанія?

1) Ми не можемо сказати, що тестова кампанія покращила кліки
2) Також немає доказів, що кампанія вплинула на пошукові запити
3) Щодо середньої кількості додавань до кошика, контрольна група показала кращий результат, тобто експериментальна кампанія спрацювала гірше 

> Експериментальна маркетингова кампанія не показала ознак ефективності. 

У двох метриках (кліки, пошукові запити) не було статистично значущої різниці, а в третій (додавання до кошика) контрольна група перевершила тестову з великою статистичною значущістю. Це означає, що експериментальна кампанія не покращила показники.
