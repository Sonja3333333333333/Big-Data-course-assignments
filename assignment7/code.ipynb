```python
import pandas as pd
from googleapiclient.discovery import build
import numpy as np
```

Використовуємо бібліотеку googleapiclient.discovery для збору інформації про 100 (може бути інша кількість) найпопулярніших відео з певної країни (не України) в Youtube 

```python
def get_the_most_popular_videos(api_key, max_results=100):
    youtube = build("youtube", "v3", developerKey=api_key)

    #список для зберігання деталей відео
    videos = []

    request = youtube.videos().list(
        part='snippet,contentDetails,statistics',
        chart='mostPopular',
        regionCode='MA',
        maxResults=100
    )

    while request and len(videos) < max_results:
        response = request.execute()
        for item in response['items']:
            video_details = {
                'video_id': item['id'],
                'title': item['snippet']['title'],
                'description': item['snippet']['description'],
                'published_at': item['snippet']['publishedAt'],
                'channel_id': item['snippet']['channelId'],
                'channel_title': item['snippet']['channelTitle'],
                'category_id': item['snippet']['categoryId'],
                'tags': item['snippet'].get('tags', []),
                'duration': item['contentDetails']['duration'],
                'definition': item['contentDetails']['definition'],
                'caption': item['contentDetails'].get('caption', 'false'),
                'view_count': item['statistics'].get('viewCount', 0),
                'like_count': item['statistics'].get('likeCount', 0),
                'dislike_count': item['statistics'].get('dislikeCount', 0),
                'favorite_count': item['statistics'].get('favoriteCount', 0),
                'comment_count': item['statistics'].get('commentCount', 0)
            }
            videos.append(video_details)



    return videos[:max_results]
```

```python
def create_csv(data, file):
    df = pd.DataFrame(data)
    df.to_csv(file)
```

### Обчисліть коефіцієнт кореляції між кількістю переглядів і кількістю вподобайок?

```python
view_count = the_most_popular_videos["view_count"]
like_count =  the_most_popular_videos["like_count"]

correlation_matrix = np.corrcoef(view_count, like_count)
correlation_coeff = correlation_matrix[0][1]
float(correlation_coeff)
```

В pandas:

```python
correlation = view_count.corr(like_count)
float(correlation)
```

Коефіцієнт кореляції між кількістю переглядів і кількістю вподобайок становить 0.83, що вказує на сильну позитивну кореляцію між цими двома змінними. Це означає, що відео з більшою кількістю переглядів, ймовірно, мають також велику кількість вподобайок.

### З якої категорії найбільше відео серед найпопулярніших?

```python
grouped_by_category = the_most_popular_videos.groupby("category_id")["video_id"].count()
grouped_by_category = grouped_by_category.rename("video_count").reset_index()
grouped_by_category
```

```python
max_value = np.max(grouped_by_category["video_count"])
row = grouped_by_category[grouped_by_category["video_count"] == max_value]
category_id = row["category_id"].iloc[0]
int(category_id)
```

Серед найпопулярніших відео категорія з ідентифікатором 22 містить найбільше відео.

### В якій категорії відео середня кількість коментарів є найбільшою?

```python
grouped_by_category = the_most_popular_videos.groupby("category_id")["comment_count"].mean()
grouped_by_category = grouped_by_category.rename("comment_count_mean_value").reset_index()
grouped_by_category
```

```python
max_value = np.max(grouped_by_category["comment_count_mean_value"])
row = grouped_by_category[grouped_by_category["comment_count_mean_value"] == max_value]
int(row["category_id"].iloc[0])
```

Середня кількість коментарів є найбільшою у категорії з ідентифікатором 20, що свідчить про вищу активність користувачів щодо відео в цій категорії.

### О котрій годині (не потрібен точний час з хвилинами) найчастіше публікують популярні відео?

```python
the_most_popular_videos["published_at"] = pd.to_datetime(the_most_popular_videos["published_at"])
the_most_popular_videos["publishing_hour"]= the_most_popular_videos["published_at"].dt.hour
grouped_by_publishing_hour = the_most_popular_videos.groupby("publishing_hour")["publishing_hour"].count().rename("frequency").reset_index()
grouped_by_publishing_hour
```

```python
max_value = np.max(grouped_by_publishing_hour["frequency"])
row = grouped_by_publishing_hour[grouped_by_publishing_hour["frequency"] == max_value]
int(row["publishing_hour"].iloc[0])
```

Популярні відео найчастіше публікуються близько 15-ї години, що вказує на тенденцію до активності в цей період для публікації відео.
